*Всемирное время* ${\rm UT}$ связано с вращением Земли. Подключим необходимые библиотеки для работы:

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

Создадим массив с данными:

````python
data = np.array([ [y, dt] for [y, dt] in [
            [-1000, 25400], [-900, 23700], [-800, 22000], [-700, 20400], [-600, 18800], [-500, 17190],
            [-400, 15530], [-300, 14080], [-200, 12790], [-100, 11640], [0, 10580], [100, 9600],
            [200, 8640], [300, 7680], [400, 6700], [500, 5710], [600, 4740], [700, 3810], [800, 2960],
            [900, 2200], [1000, 1570], [1100, 1090], [1200, 740], [1300, 490], [1400, 320], [1500, 200],
            [1600, 120], [1700, 9], [1710, 10], [1720, 11], [1730, 11], [1740, 12], [1750, 13],
            [1760, 15], [1770, 16], [1780, 17], [1790, 17], [1800, 14], [1810, 13], [1820, 12],
            [1830, 8], [1840, 6], [1850, 7], [1860, 8], [1870, 2], [1880, -5], [1890, -6], 
            [1900, -3], [1910, 10], [1920, 21], [1930, 24], [1940, 24], [1950, 29], [1960, 33],
            [1970, 40], [1980, 51], [1990, 57], [2000, 65]]])
````

Изобразим все данные на графике:

````python
plt.plot(data[:,0], data[:,1], 'g')
plt.title('Данные за всё время')
plt.xlabel('Год')
plt.ylabel('$\Delta T$, сек')
plt.grid()
plt.show()
````

Теперь построим график для данных, начиная с 1700 года:

````python
data_recent = data[np.where(data[:,0]>=1700)]
plt.plot(data_recent[:,0], data_recent[:,1], 'g')
plt.title('Данные, начиная с 1700 г.')
plt.xlabel('Год')
plt.ylabel('$\Delta T$, сек')
plt.grid()
plt.show()
````

Подберём многочлены степеней 1-4, аппроксимирующие данные. Для этого используем метод наименьших квадратов.
Как было сказано в задании, сначала установим год для сдвинутой шкалы по времени (1800 г.):

````python
shift_year = 1800
````

Отделим данные, начиная с 1800 года:

````python
data_after_1800 = data[data[:, 0] >= shift_year]
````

Затем создадим функцию для нахождения полинома:

````python
def fit_data(data, degree):
  x, y = data[:, 0], data[:, 1]
  x_shifted = x - shift_year                 # сдвигаем шкалу
  coeffs = np.polyfit(x_shifted, y, degree)  # находим коэффициенты многочлена
  poly = np.poly1d(coeffs)                   # создаём многочлен
  return poly
````

Также содадим функцию для вычисления "классического приближения" по следующей формуле:
$$
32\Bigl(({\rm year} - 1820)/100\Bigr)^2
$$

````python
def classical_approx(years):
  return 32 * ((years - 1820) / 100) ** 2
````

Зададим степени и сгенерируем многочлены:

````python
degrees = [1, 2, 3, 4]
polynomials = [fit_data(data_after_1800, degree) for degree in degrees]
````

Выведем полученные многочлены на экран:


````python
counter = 1
for p in polynomials:
  print(f'Многочлен {counter} степени:\n', p)
  if counter != 4: print('\n')
  counter += 1
````

Построим график сравнения аппроксимаций и "классического" приближения:

````python
plt.figure(figsize=(14, 5))
years_shifted = data_after_1800[:, 0] - shift_year
plt.plot(data_after_1800[:, 0], data_after_1800[:, 1], 'ro', label='Данные')
plt.plot(data_after_1800[:, 0], classical_approx(data_after_1800[:, 0]), label='Классическое приближение')
for poly, degree in zip(polynomials, degrees):
    plt.plot(data_after_1800[:, 0], poly(years_shifted), label=f'Степень {degree}')
plt.xlabel('Год')
plt.ylabel('$\Delta T$, сек')
plt.legend()
plt.title('Сравнение аппроксимаций и "классического" приближения')
plt.grid()
plt.show()
````

##**Вывод**:
Смотря на график, можно сделать вывод, что многочлены степени 2 и выше достаточно хорошо аппроксимируют данные, при этом результаты многочленов 2-й, 3-й и 4-й степени почти совпадают друг с другом, что указывает на то, что, вероятно, последующее дополнительное увеличение степени почти не улучшает аппроксимацию. "Классическое" приближение также неплохо аппроксимирует данные, но не даёт такой точности, как многочлены степени 2-4.
